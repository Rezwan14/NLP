```python
# imports

import urllib.request
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import RegexpTokenizer
```

```python
# load the data
shakespeare_url = "http://www.cs.columbia.edu/~sarahita/CL/lab1/shakespeare.txt"
news_url = "http://www.cs.columbia.edu/~sarahita/CL/lab1/news.txt"
swbd_url = "http://www.cs.columbia.edu/~sarahita/CL/lab1/swbd.txt"

# read url .txt file into string "data"
def get_data(url):
  data = urllib.request.urlopen(url).read().decode('utf-8')
  return data

shakespeare_data = get_data(shakespeare_url)
news_data = get_data(news_url)
swbd_data = get_data(swbd_url)
```

```python
#find number of words in news_data
news_words = word_tokenize(news_data)
len(news_words)
```

```python
#words in shakesphere
shake_words = word_tokenize(shakespeare_data)
len(shake_words)
```

```python
#number of unique words in news_data
news_word_dist = nltk.FreqDist(news_words)
Unique_NewsWords = dict((word, freq) for word, freq in news_word_dist.items() if not word.isdigit())

#print(Unique_NewsWords)
len(Unique_NewsWords)
```

```python
#number of unique words in swbd_data
swbd_word_dist = nltk.FreqDist(swbd_words)
Unique_SwbdWords = dict((word, freq) for word, freq in swbd_word_dist.items() if not word.isdigit())

print(Unique_SwbdWords)
len(Unique_SwbdWords)
```

```python
#number of unique words in shakesphere_data
shake_word_dist = nltk.FreqDist(shake_words)
Unique_ShakeWords = dict((word, freq) for word, freq in shake_word_dist.items() if not word.isdigit())

print(Unique_ShakeWords)
len(Unique_ShakeWords)
```

```python
#Type to token ratio (TTR) V/N for shakesphere_data
Shake_TTR = (len(Unique_ShakeWords) / len(shake_words))
Shake_TTR
```

```python
#remove punctuation for news
import string
regular_punct = list(string.punctuation)
def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()

news_punct = remove_punctuation(news_lower,regular_punct)
news_punct
```

```python
#remove punctuation for shakesphere
regular_punct = list(string.punctuation)
def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()

shake_punct = remove_punctuation(shake_lower,regular_punct)
shake_punct
```

```python
#remove punctuation for swbd
import string
regular_punct = list(string.punctuation)
def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()

swbd_punct = remove_punctuation(swbd_lower,regular_punct)
swbd_punct
```

```python
#remove stopwords in news text
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words("english")
news_stop = nltk.word_tokenize(news_punct)
news_stopwords = [word for word in news_stop if word not in stop_words]
print(news_stopwords)
```

```python
#remove stopwords in swbd text
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
swbd_stop = nltk.word_tokenize(swbd_punct)
swbd_stopwords = [word for word in swbd_stop if word not in stop_words]
print (swbd_stopwords)
```
